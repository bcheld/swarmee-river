## Some notebook context

The magic includes the notebook context (markdown + code) in its prompt, excluding images.
Add a few cells of code and then ask Swarmee about them.

````python
from dataclasses import dataclass


@dataclass
class Item:
    name: str
    price_cents: int


def total_cost(items: list[Item]) -> float:
    """Return total cost in dollars."""
    return sum(i.price_cents for i in items) / 100.0


items = [Item("apple", 199), Item("banana", 79)]
total_cost(items)
````

````python
%%swarmee
Given the notebook context, suggest improvements to `total_cost` (types, naming, edge cases).
````
